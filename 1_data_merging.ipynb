```python
import pandas as pd
import numpy as np
```

```python
DATA_PATH = "../../data/BR_env/"
FILES = [
    '1610.xlsx',
    '1611.xlsx',
    '1612.xlsx',
    '1701.xlsx',
    '1702.xlsx',
    '1703.xlsx',
    '1704.xlsx',
    '1705.xlsx',
    '1706.xlsx',
    '1707.xlsx',
    '1708.xlsx',
    '1709.xlsx',
    '1710.xlsx',
    '1711.xlsx',
    '1712.xlsx',
    '1801.xlsx',
    '1802.xlsx',
    '1803.xlsx',
    '1804.xlsx',
    '1805.xlsx'
]
COLUMNS = [
    'N1 온도', 'N1 습도', 'N1 지온', 'N1 지습', 'N1 EC', 'N1 CO2', 'N1 대기압', 'N1 UV', 'N1 조도',
    'N2 온도', 'N2 습도', 'N2 CO2', 'N2 대기압', 'N2 UV', 'N2 조도',
    'N3 온도', 'N3 습도', 'N3 지온', 'N3 지습', 'N3 EC', 'N3 CO2', 'N3 대기압', 'N3 UV', 'N3 조도',
    'N4 온도', 'N4 습도', 'N4 CO2', 'N4 대기압', 'N4 UV', 'N4 조도',
    'N5 온도', 'N5 습도', 'N5 지온', 'N5 지습', 'N5 EC', 'N5 CO2', 'N5 대기압', 'N5 UV', 'N5 조도',
    'N6 온도', 'N6 습도', 'N6 지온', 'N6 지습', 'N6 EC', 'N6 CO2', 'N6 대기압', 'N6 UV', 'N6 조도',
    'N7 온도', 'N7 습도', 'N7 지온', 'N7 지습', 'N7 EC', 'N7 CO2', 'N7 대기압', 'N7 UV', 'N7 조도',
    'N8 온도', 'N8 습도', 'N8 CO2', 'N8 대기압', 'N8 UV', 'N8 조도',
    'N9 온도', 'N9 습도', 'N9 지온', 'N9 지습', 'N9 EC', 'N9 CO2', 'N9 대기압', 'N9 UV', 'N9 조도',
    '기상센서\n온도', '기상센서\n습도', '기상센서\nUV', '기상센서\n풍향', '기상센서\n풍속', '기상센서\n대기압', '기상센서\n강우']

DATE_RANGE = pd.date_range('2016-10-01 00:00', '2018-05-31 23:50', freq='10min')
```

```python
df_list = []
for FILE_NAME in FILES:
    print("%s data loading... " % FILE_NAME, end = '')
    df = pd.read_excel(DATA_PATH+FILE_NAME, skiprows=6, encoding="utf-8", index_col="N1 날짜")
    df = df.loc[df.index.to_series().dropna()]
    df = df[COLUMNS]
    df = df[df.index != '===============']
    df.index = pd.DatetimeIndex(df.index)
    df.index = df.index.round(freq='10min')
    df_list.append(df)
    print("OK")

merged_df = pd.concat(df_list)
```

```python
merged_df = merged_df.reindex(DATE_RANGE)
merged_df = merged_df[~merged_df.index.duplicated()]
merged_df.info()
```
